```python
df = df.drop(['Curricular units 2nd sem (credited)','Curricular units 2nd sem (without evaluations)','Curricular units 1st sem (without evaluations)'], axis = 1)
```

```python
df.rename(columns = {'Curricular units 2nd sem (without evaluations)':'CU_2nd_sem_without evaluations'}, inplace = True)
```

Visualize the Data

```python
# Bar plot for target variable distribution
plt.figure(figsize=(6, 6))
sns.countplot(data=df, x='Target',edgecolor='black')
plt.title('Distribution of Academic Success')
plt.xlabel('Academic Success')
plt.ylabel('Count')
plt.show();
```

```python
plt.bar(list(df['Gender'].value_counts().index),df['Gender'].value_counts())
```

```python
plt.scatter(df['Gender'],df['Course']);
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
plt.figure(figsize = (15, 9))
sns.lineplot(x = 'Application order',y ='Marital status', data = df, marker='o',mfc = 'red', color = 'black', ls = '--',
        ms = 20) 
plt.title('Line plot')
plt.xlabel(' Application order')   # mfc - marker fonter  # ms= marker size   # ls - line style  # gride is 
plt.ylabel(' Marital status')

plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.countplot(data=df, x='Course',edgecolor='black')
plt.title('Distribution of Academic Success')
plt.xlabel('Course')
plt.ylabel('Count')
plt.xticks(range(1, 20, 5))
plt.show();
```

```python
sns.jointplot(x = 'Age at enrollment', y = 'GDP', data = df,
             hue = 'Gender');
```
