# Checking for missing values

'df1.isnull().sum()'  is a way to count the number of missing values in each column of a pandas DataFrame df1.

```python
df1.isnull().sum()
```

No null values present in our dataset, but we will drop unamed column because it does not holds any values.

```python
df1.drop(columns=['Unnamed: 0'], inplace=True)
df1.head()
```

**Here, we have to create Function to remove dollars and other symbols as they will create issues during plotting graphs**

```python
# create function
def clean_data(column):
    return pd.to_numeric(column.replace('[\$,B%,]', '', regex=True))

# Clean and convert the columns
df1['GNI'] = clean_data(df1['GNI'])
df1['Per_Capita'] = clean_data(df1['Per_Capita'])
df1['Annual_Change'] = clean_data(df1['Annual_Change'])
```

**Describe the dataset**

'describe()' method of pandas shows all the statistical values of the dataset like count, mean, standard deviation, minimum, maximum, etc.

```python
df1.describe()
```

# Data Visualization

**So, here we are ploting 'lineplot' for finding the GNI (Gross national income) of India Year on Year.**

```python
#Lets start with analysis
#Growth over time on yearly basis
plt.figure(figsize=(15,9))
sns.lineplot(x='year', y='GNI', data=df1, marker='o')
plt.title("GNI of India Year-on-Year")
plt.xlabel('Year')
plt.ylabel('GNI (in billions)')
plt.grid(True)
plt.show()
```

**We are creating 'lineplot' for GNI Per capita of India Year-on-Year.**

Where, Per Capita stands for the average per person.

```python
#Lets analyse per capita GNI
plt.figure(figsize=(15,9))
sns.lineplot(x='year', y='Per_Capita', data=df1, marker='o', color='green')
plt.title("GNI Per capita of India Year-on-Year")
plt.xlabel('Year')
plt.ylabel('GNI (in dollars)')
plt.grid(True)
plt.show()
```

# Checking the distribution of the Academic Success  over the GNI

```python
plt.figure(figsize=(6, 6))
sns.countplot(data=df1, x='GNI',edgecolor='black')
plt.title('Distribution of Academic Success')
plt.xlabel('GNI')
plt.ylabel('Count')
plt.xticks(range(1, 50, 5))
plt.show();
```

**This plot can help visualize the relationship between GNI and Annual_Change over time, with the color of each point indicating the year.**

```python
plt.scatter(df1['GNI'], df1['Annual_Change'], c = df1['year']);
```
